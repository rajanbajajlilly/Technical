# Inner Joins 

Merges 2 datasets based on common values

```python
week1.merge(week2, how="inner", on="Customer ID")
```

```python
 week1[week1["Customer ID"] == 155]
```

```python
week1.merge(week2, how="inner", on="Customer ID", suffixes = [" - Week 1", " - Week 2"])
```

```python
week1.merge(week2, how = "inner", on = ["Customer ID", "Food ID"])
```

```python
week1[week1["Customer ID"] == 21]
```

```python
week2[week2["Customer ID"] == 21] #matched twice
```

# Outer Joins 

```python
merged = week1.merge(week2, how = "outer", on = ["Customer ID"], suffixes = [" - Week1", " - Week2"], 
            indicator = True)
```

```python
week1.merge(week2, how = "outer", on = ["Customer ID", "Food ID"], suffixes = [" - Week1", " - Week2"])
```

```python
merged["_merge"].value_counts()    
```

```python
mask = merged["_merge"].isin(["left_only", "right_only"])
merged[mask]
```

```python
mask1 = merged["_merge"].isin(["both"])
merged[mask1]
```

# Left Joins

```python
mask3 = merged["_merge"].isin(["left_only", "both"])
merged[mask3]
```

# The left_on and right_on parameters

When the column names are different in both dataframes

```python
week2.head(3)
```

```python
customers.head(3)
```

```python
week2.merge(customers, how = "left", left_on = "Customer ID", right_on = "ID")
```

```python
week2.merge(customers, how = "left", left_on = "Customer ID", right_on = "ID").drop("ID", axis = "columns")
```

```python
week2.merge(customers, how = "left", left_on = "Customer ID", right_on = "ID", sort = "Customer ID").drop("ID", axis = "columns")
```
