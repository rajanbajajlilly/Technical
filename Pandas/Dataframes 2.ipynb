Filter with m ore than one condition - AND

```python
mask1 = df['Gender'] == 'Male'
mask2 = df['Team'] == 'Marketing'

df[mask1 & mask2]
```

Filter with more than one condition (OR)

```python
mask1 = df['Senior Management'] == True
mask2 = df['Start Date'] < '1990-01-01'
df[mask1 | mask2]
```

```python
mask1 = df['First Name'] == 'Robert'
mask2 = df['Team'] == 'Client Services'
mask3 = df['Start Date'] > '2016-06-01'
df[(mask1 & mask2) | mask3]
```

The .isin() method

```python
mask1 = df['Team'] == 'Legal'
mask2 = df['Team'] == 'Sales'
mask3 = df['Team'] == 'Product'

df[mask1 | mask2 | mask3]
```

```python
df['Team'].isin(['Legal', 'Sales', 'Product'])
```

```python
mask = df['Team'].isin(['Legal', 'Sales', 'Product', 'Marketing'])
df[mask]
```

The .duplicated() method

```python
df.sort_values('First Name', inplace = True)
```

```python
df['First Name']
```

```python
df['First Name'].duplicated() #First element found is marked as unique, rest duplicate
```

```python
df[df['First Name'].duplicated()]
```

```python
df[df['First Name'].duplicated(keep = 'last')]
```

```python
df[df['First Name'].duplicated(keep = False)] #mark duplicated all after the first occurence
```

```python
~df['First Name'].duplicated(keep = False)
```

```python
mask = ~df['First Name'].duplicated(keep = False)
df[mask]
```

The .drop_duplicates() method

```python
df.head(3)
```

```python
len(df.drop_duplicates())
```

```python
df.drop_duplicates(subset = ['First Name'], keep = 'first') #subset repr name of the columns to check for duplicate
```

```python
df.drop_duplicates(subset = ['First Name'], keep = False) #removes all duplicates for first name
```

```python
df.drop_duplicates(subset = ['Team'], keep = False) #no single team values taht occur only once
```

```python
df.drop_duplicates(subset = ['First Name', 'Team']) #both columns must be duplicate to get dropped
```
